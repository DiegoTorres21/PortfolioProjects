```python
#Import libraries

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plots we will create

```

```python
#Read in the data
df = pd.read_csv(r'movies.csv')
```

```python
#Let's look at the data
df.head()
```

```python
#Let's see if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(f'{col} - {pct_missing*100}%')
    
```

```python
pd.set_option('display.max_rows', None)

# Get the rows where any of the columns have null values
# Display the rows as a DataFrame
rows_with_null = df[df.isnull().any(axis=1)]


# Display the styled DataFrame with frozen headers
rows_with_null_style = rows_with_null.style.set_table_styles([
    {'selector': 'th',
     'props': [('position', 'sticky'),
               ('top', '0'),
               ('background-color', 'white')]}
])



rows_with_null_style
```

```python
num_rows = len(df)
print("Number of rows:", num_rows)
num_rows_with_nulls = len(rows_with_null )
print("Number of rows:", num_rows_with_nulls)
```

```python
#Drop rows with missing data

df = df.dropna()
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(f'{col} - {pct_missing*100}%')
```

```python
#Data types for our columns

df.dtypes
```

```python
#Change datatype of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64') 
df.head()
```

```python
#Create correct year column  
df['yearcorrect'] = df['released'].astype(str).str.split(',').str[1].str[1:5]
df.head()
```

```python
df = df.sort_values(by=['gross'],  inplace = False, ascending = False)
```

```python
#Drop any duplicates
df.drop_duplicates()
df.head()
```

```python
#Budget high correlation
#company high correlation
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
#Plot budget vs gross using seaborn 
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws={"color":"red"}, line_kws = {"color" : "blue"}) 
```

```python
#Let's start looking at correlation
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features');
```

```python
# Looks at company
df.head()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes 
        
df_numerized.head()
            
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matrix for categoric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features'); 
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending = False)
sorted_pairs
```

```python
high_corr =  sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
